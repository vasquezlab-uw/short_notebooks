### compare denoised in Elements and in macro
we have a have a macro that runs denoised.ai on all the files in a directory, but we are unsure whether it runs the same process

```python
# import packages
from skimage.io import imread, imshow
import matplotlib.pyplot as plt
```

```python
# import images to compare
# notebook must be in same directory as images OR need to specify path
im1 = imread("231003_cutb_x_ommmcherry_live_018 - Denoised.nd2.tif")[:,0, :, :]
im2 =imread("231003_cutb_x_ommmcherry_live_018 - Denoised_notMacro.tif")[:, 0, :, :]
```

```python
# show max z projection of each image
fig, ax = plt.subplots(1, 2)
ax[0].imshow(im1.max(0))
ax[0].set_title('denoised, macro')
ax[1].imshow(im2.max(0))
ax[1].set_title('denoised, not macro')
```
